```python
from bs4 import BeautifulSoup
import datetime
import glob
import os
import pandas as pd

%matplotlib inline
```

```python
def get_df_from_html_path(path):
    """ We read the historical scraped html
        to get the prices and marketcaps of the top 200 coins.
        We return a pandas dataframe with this information.
    """
    with open(path, "rb") as f:
        soup = BeautifulSoup(f, "html.parser")
        list_of_lists = []
        table_rows = soup.find_all("tr")
        table_header = table_rows[0].getText(separator="||").split("||")
        for tr in table_rows[3:]:
            list_of_lists.append(tr.getText(separator="||").split("||"))
#     print(table_header)
#     print(list_of_lists[0])
    dataframe = pd.DataFrame(list_of_lists, columns=table_header)
    ymd = datetime.date(int(path[-13:-9]), int(path[-9:-7]), int(path[-7:-5]))
    dataframe["date"] = ymd
    return dataframe
```

```python
# here we are grabbing all of the files that we want by globbing them and reading them all in
df_list = []
for html_name in glob.glob("./coindesk/coinmarketcap-historical-*.html"):
#     print(html_name)
    try:
        df = get_df_from_html_path(html_name)
        print(html_name)
    except:
        print(f"could not process: {html_name}")
    df_list.append(df)
```

```python
def dollar_str_to_float(dollar_str):
    if pd.notnull(dollar_str):
        return float(dollar_str[1:].replace(",", ""))
    else:
        return dollar_str
```

```python
historical_df = pd.concat(df_list )
historical_df["market_cap"] = historical_df["Market Cap"].apply(dollar_str_to_float)
historical_df["price"] = historical_df["Price"].apply(dollar_str_to_float)
```

```python
def get_df_from_html_path_today(path):
    """ Here we read the prices for today's cryptocurrencies
        (all of them, not just the top 200). They're stored in a slightly different
        format so we have a different function.
    """
    table_header = ["Rank", "Name", "Rank_Again", "Symbol",
                    "Price", "24h", "7d", "Market Cap",
                    "Volume USD", "Volume Units" ,"Circulating Supply"]
    reduced_header = ["Name", "Symbol", "Price", "24h", "7d"]
    with open(path, "rb") as f:
        soup = BeautifulSoup(f, "html.parser")
        list_of_lists = []
        list_of_reduced_lists = []
        table_rows = soup.find_all("tr")
        for tr in table_rows[1:]:
            row_data = tr.getText(separator="||").split("||")
            if len(row_data) == len(table_header):
                list_of_lists.append(row_data)
            elif len(row_data) == 5:
                list_of_reduced_lists.append(row_data)
            else:
                print(row_data)
#                 print(row_data)
#     print(table_header)
#     print(list_of_lists[0])
    dataframe = pd.DataFrame(list_of_lists, columns=table_header)
    reduced_dataframe = pd.DataFrame(list_of_reduced_lists, columns=reduced_header)
    return pd.concat([dataframe, reduced_dataframe], sort=False)
```

```python
df_list = []
for html_name in glob.glob("./coindesk/coinmarketcap-current-*-20210103.html"):
#     print(html_name)
    df = get_df_from_html_path_today(html_name)
    df_list.append(df)
```

```python
todays_price_df = pd.concat(df_list)
```

```python
todays_price_df["price"] = todays_price_df["Price"].apply(dollar_str_to_float)
todays_price_df["market_cap"] = todays_price_df["Market Cap"].apply(dollar_str_to_float)
todays_price_df["volume_usd"] = todays_price_df["Volume USD"].apply(dollar_str_to_float)
```

```python
todays_price_df.reset_index(drop=True, inplace=True)
```

```python
todays_price_df.to_csv("./coinmarketcap_current_price.csv", index=False)
```
